```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import Dataset
iris = sns.load_dataset('iris')
iris.head(5)
```

```python
#Basic plot - scatter plot between sepal_length and sepal_width
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue = 'species')
plt.title('Scatter Plot between sepal_length and sepal_width')
plt.show()
```

```python
#Categorical - bar chart
sns.barplot(data=iris, x='sepal_length', y='sepal_width', hue = 'species')
plt.title('average sepal_length by sepal_width')
plt.show()
```

```python
sns.countplot(data=iris, x='species')
plt.xlabel('species')
plt.ylabel('Count')
plt.title('Count of Species')
plt.show()
```

```python
sns.boxplot(x='species', y='petal_length', data=iris)
plt.show()
```

```python

corr =iris.drop(columns=['species']).corr()
sns.heatmap(corr, annot = True, cmap = 'coolwarm')
plt.title('HeatMap')
plt.show()
```
