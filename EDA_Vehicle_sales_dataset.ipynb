**Identify Trends in Vehicle Sales Over Time**

```python
# Inspect the 'saledate' column
print(df['saledate'].head())
```

```python
# Identify invalid date rows by using 'errors' parameter in pd.to_datetime()
invalid_dates = pd.to_datetime(df['saledate'], errors='coerce')  # Converts invalid dates to NaT (Not a Time)

# Check rows where the date couldn't be parsed
invalid_rows = df[invalid_dates.isna()]
print(invalid_rows[['saledate', 'make', 'model']])
```

```python

df['saledate'] = df['saledate'].replace('16500', '1650')
```

```python
# Drop rows where 'saledate' couldn't be parsed
df = df[invalid_dates.notna()]
```

```python
# Convert 'saledate' to datetime format after cleaning
df['saledate'] = pd.to_datetime(df['saledate'], errors='raise')  # Use 'raise' now to catch any remaining issues
```

```python
# Check if any timezone information is present in the saledate column
print(df['saledate'].apply(lambda x: x.tzinfo is not None).value_counts())
```

```python
# Convert timezone-aware datetimes to UTC
df['saledate'] = pd.to_datetime(df['saledate'], utc=True)
```

```python
# Remove timezone information (convert to naive datetime)
df['saledate'] = pd.to_datetime(df['saledate']).dt.tz_localize(None)
```

```python
# Now, proceed with the date analysis
df['year_month'] = df['saledate'].dt.to_period('M')

# Group by year and month to get the number of sales
sales_trend = df.groupby('year_month').size()

# Plot the trend over time
plt.figure(figsize=(10,6))
sales_trend.plot(kind='line', color='blue')
plt.title('Vehicle Sales Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Number of Vehicles Sold')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python

# Scatter plot of odometer vs selling price
plt.figure(figsize=(16,8))
sns.scatterplot(x='odometer', y='sellingprice', data=df)
plt.title('Odometer vs Selling Price')
plt.xlabel('Odometer (Mileage)')
plt.ylabel('Selling Price')
plt.grid(True)
plt.show()



```

```python
# Boxplot to visualize the effect of condition on selling price
plt.figure(figsize=(16,8))
sns.boxplot(x='condition', y='sellingprice', data=df)
plt.title('Condition vs Selling Price')
plt.xlabel('Condition')
plt.ylabel('Selling Price')
plt.grid(True)
plt.show()
```

```python
# Scatter plot for MMR values vs actual selling price
plt.figure(figsize=(8,6))
sns.scatterplot(x='mmr', y='sellingprice', data=df)
plt.title('MMR Values vs Selling Prices')
plt.xlabel('MMR Value')
plt.ylabel('Selling Price')
plt.grid(True)
plt.show()

# Calculate correlation between MMR and Selling Price
correlation = df['mmr'].corr(df['sellingprice'])
print(f"Correlation between MMR and Selling Price: {correlation}")

```

```python
# Count the number of sales per make
top_makes = df['make'].value_counts().head(10)
print(top_makes)



```

```python
# Plot top 10 vehicle makes
plt.figure(figsize=(10,6))
top_makes.plot(kind='bar', color='skyblue')
plt.title('Top 10 Vehicle Makes by Sales Volume')
plt.xlabel('Vehicle Make')
plt.ylabel('Number of Vehicles Sold')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Count the number of sales per model
top_models = df['model'].value_counts().head(10)
print(top_models)

# Plot top 10 vehicle models
plt.figure(figsize=(10,6))
top_models.plot(kind='bar', color='blue')
plt.title('Top 10 Vehicle Models by Sales Volume')
plt.xlabel('Vehicle Model')
plt.ylabel('Number of Vehicles Sold')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
