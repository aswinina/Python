```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import Dataset
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
#Basic plot - scatter plot between total bill and tips
sns.scatterplot(data=tips, x='total_bill', y='tip', hue = 'time')
plt.title('Scatter Plot between Total Bill and Tips')
plt.show()
```

```python
#Categorical - bar chart
sns.barplot(data=tips, x='day', y='total_bill', hue = 'time')
plt.title('Average bill by day')
plt.show()
```

```python
#Categorical - bar chart
sns.countplot(data=tips, x='day', palette='coolwarm')
plt.title('count of days')
plt.show()
```

```python
##Categorical - box chart
sns.boxplot(data=tips, x='day', y='total_bill', palette = 'Set2')
plt.title('Box plot os total Bill by Day by day')
plt.show()
```
