```python
#Import Library
import matplotlib.pyplot as plt
```

```python
#Basic Line chart

days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
temp = [23,15,14,24,27,22,21]

plt.plot(days,temp)
plt.title('Daily Temperature by week')
plt.xlabel('Days')
plt.ylabel('Temperatures')
plt.show()
```

```python
#Customise Line chart

plt.plot(days,temp,linestyle='--',marker = 'X',color='brown')
plt.title('Daily Temperature by week')
plt.xlabel('Days')

plt.ylabel('Temperatures')
plt.show()
```

```python
# Sales Data Analysis Bar chart

products = ['Product A','Product B','Product C']
sales =[100,200,300]
plt.bar(products,sales,color = 'blue')
plt.title('Sales Data Analysis')
plt.xlabel('Products')

plt.ylabel('Sales')
plt.show()
```

```python
# Market Share Analysis -Pie chart

products = ['Product A','Product B','Product C']
sales =[45,20,50]
plt.pie(sales,labels=products, autopct='%1.1f%%')
plt.title('Market shsre Distubution Analysis')
plt.show()
```

```python
#Analyze Exam score -Historgram
scores=[33,55,44,66,55,33,55,44,65,45,76,43,67,54,34,44,65]

plt.hist(scores,bins=5, color='yellow',edgecolor='black')
plt.title('Exam score Analysis')
plt.xlabel('Scores Range')

plt.ylabel('no. of Students')
plt.show()
```

```python
#Analysis between Hieght and weight - Scatter plot
height =[155,160,159,164,160,167,159,160]
weight = [61, 63, 59, 57, 55, 54, 67, 63]
plt.scatter(height,weight,color = 'red')
plt.title('Height and Weight Analysis')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```
